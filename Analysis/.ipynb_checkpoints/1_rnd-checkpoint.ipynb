```python
from gammapy.modeling.models.IRF import IRFModel
from gammapy.irf import EDispKernel


class MyERecoIRFModel(IRFModel):
    bias = Parameter("bias", "0")
    resolution = Parameter("resolution", "0")

    tag = ["ERecoIRFModel", "ereco"]
    _type = "e_reco_model"

    @staticmethod
    def evaluate(energy_axis, bias, resolution):

        gaussian = EDispKernel.from_gauss(
            energy_axis_true=energy_axis.copy(name="energy_true"),
            energy_axis=energy_axis,
            sigma=(1e-12 + np.abs(resolution.value)),
            bias=bias.value,
        )
        return gaussian
```

```python
from gammapy.irf import EDispKernelMap, EDispMap, PSFKernel, PSFMap, RecoPSFMap

def edisp_helper(ds, energy):
    energy_rebins = MapAxis(
        nodes=np.logspace(
            np.log10(energy.center[0].value),
            np.log10(energy.center[-1].value),
            self.e_reco_n * len(energy.center),
        ),
        node_type="center",
        name="energy",
        unit="TeV",
        interp="log",
    )
    return energy_rebins

def npred_edisp(ds):
    """Predicted edisp map
    Returns
    -------
    irf : `Map`
    """
    # print("npred_edisp")
    edisp = ds.edisp
    # get the kernel
    edisp_kernel = edisp.get_edisp_kernel()
    # rebin enenergyaxis
    energy_rebins = ds.edisp_helper(edisp_kernel.axes["energy"])
    # compute gaussian with new eaxis
    gaussian = ds.irf_model.e_reco_model(
        energy_axis=energy_rebins,
    )
    # rebin edisp_kernel data and multiply with gaussian
    data_rebinned = np.matmul(
        np.repeat(
            np.repeat(edisp_kernel.data, ds.e_reco_n, axis=0),
            ds.e_reco_n,
            axis=1,
        ),
        gaussian.data,
    )
    # set as kernel data
    edisp_kernel.data = data_rebinned.reshape(
        (
            len(edisp_kernel.axes["energy_true"].center),
            ds.e_reco_n,
            len(edisp_kernel.axes["energy"].center),
            ds.e_reco_n,
        )
    ).mean(axis=(1, 3))

    edisp = EDispKernelMap.from_edisp_kernel(edisp_kernel)
    return edisp
```
