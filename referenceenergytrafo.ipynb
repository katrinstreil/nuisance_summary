```python
from gammapy.modeling.models import PowerLawNormSpectralModel
import astropy.units as u
import matplotlib.pyplot as plt
```

```python
a = 0.1
beta = 0.02
f  = PowerLawNormSpectralModel(norm = a, tilt = beta, reference = 100 *u.TeV)
f.plot(label = 'Eref = 100TeV')


c = 0.1 * (1/100) ** (-beta)
print(c)
d = 0.02
f  = PowerLawNormSpectralModel(norm = c, tilt = d, reference = 1 *u.TeV)
f.plot(label = 'Eref = 1TeV')

plt.legend()
```
